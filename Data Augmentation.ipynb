```python
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
```

```python
img=image.load_img('cat.825.jpg',target_size=(150,150))
plt.imshow(img)
```

```python
datagen=ImageDataGenerator(rotation_range=30,shear_range=0.2,zoom_range=0.2,horizontal_flip=True,fill_mode='constant',
                               width_shift_range=0.2,height_shift_range=0.2)
```

```python
img=image.img_to_array(img)
img.shape
```

```python
inp_batch=img.reshape(1, 150, 150, 3)
```

```python
i=0
for op in datagen.flow(inp_batch,batch_size=1,save_to_dir='aug'):
    i+=1
    if i==10:
        break
```
